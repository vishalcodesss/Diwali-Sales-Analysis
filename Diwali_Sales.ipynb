```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('./Diwali Sales Data.csv', encoding='unicode_escape')
# To avoid encoding error, encoding='unicode_escape' is used 
```

```python
df.shape    # to get the shape or dimensions of the dataframe
```

```python
df.info()   # to get the info about the dataframe
```

```python
df.drop(['Status','unnamed1'], axis=1, inplace=True)
# Drop the columns Status and unnamed1 permanently
# axis=1 means total column
```

```python
df.info()
```

```python
df.isna().sum()     # Finds the count of null values in the columns
```

```python
df.dropna(inplace=True)     # Drop null rows permanently
```

```python
df['Amount'] = df['Amount'].astype('int')
# Used to change the datatypes of the columns 
# Like in this case, Amount column was float type and was converted to int
```

```python
df['Amount'].dtypes
#  Used to see the datatypes of the columns
```

Exploratory Data Analysis

```python
ax = sns.countplot(x='Gender', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Gender_sales = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False)
```

```python
sns.barplot(x='Gender', y='Amount', data=Gender_sales)
```

From the above graphs we can see that most of the buyers are females and the purchasing power of females are greater than men

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False)
# The as_index returns a pandas dataframe, hence we can use sort_values with by parameter to sort the dataframe.
# If we don't use as_index=false then, it will return as a series and then we cannot use 'by' parameter in sort_values
```

Age

```python
ay = sns.countplot(x='Age Group', data=df, hue='Gender')

for bars in ay.containers:
    ay.bar_label(bars)      # this block of code is used to show data labels in the graph
```

```python
age_vs_sales = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group', y='Amount', data=age_vs_sales)
```

From the above graphs, we can see that most of the buyers are of age group between 26-35 yrs and gender is mostly female

State

```python
df.columns
```

```python
# Total Number of Orders from top 10 states
orders_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})      # this is used to set the size of the figure 
sns.barplot(data=orders_state, x='State', y='Orders')
```

```python
# Total sales from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})   
sns.barplot(x='State', y='Amount', data=sales_state)
```

From the above graphs, we can see that most of the orders and total sales/amount are from Uttar Pradesh, Maharastra and Karnataka respectively

Marital Status

```python
az = sns.countplot(data=df, x='Marital_Status')

sns.set(rc={'figure.figsize':(5,5)})  
for bars in az.containers:
    az.bar_label(bars)
```

```python
sales_marital = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(data=sales_marital, x= 'Marital_Status', y='Amount', hue='Gender')
```

From the above graphs, we saw that most of the buyers are married women and they have high purchasing power

Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})  
aa = sns.countplot(data=df, x='Occupation')

for bars in aa.containers:
    aa.bar_label(bars)
```

```python
occupation_sales = df.groupby(['Occupation', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(data=occupation_sales, x='Occupation', y='Amount', hue='Gender')
```

From the above graph, we can see that most of the buyers are working in IT, Aviation and Healthcare

Product Category

```python
ax = sns.countplot(data=df, x='Product_Category')
sns.set(rc={'figure.figsize':(20,8)})  

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
category_sales = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize': (40,15)})
sns.barplot(data=category_sales, x='Product_Category', y='Amount')
```

From the above graphs, we can see that most of the sold products are from food, clothing and electronics category
